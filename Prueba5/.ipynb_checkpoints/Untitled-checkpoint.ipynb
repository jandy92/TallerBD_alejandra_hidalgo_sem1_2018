En un histograma, visualizar el GDP per Capita de cada país. Bonus: ordenar el histograma en orden decreciente.




```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('dictionary.csv', sep=",")

df[:3]
```

```python
r=df.groupby('Country').mean().plot(kind='bar',stacked=True,figsize=(15,10))
plt.show()
```

¿Cuál es el número de medalla por cada país, sin diferenciar el tipo de medalla, en los juegos de verano? Visualizar en un histograma.



```python
s = pd.read_csv('summer.csv', sep=",")

s[:3]
```

```python
med=s['Medal']
city=s['City']
ctmedal=med.str.contains('')
citymetal=pd.concat([city, ctmedal], axis=1)
citymetal.groupby('City').aggregate(sum).plot(kind="bar",stacked=True)
plt.show()
```
